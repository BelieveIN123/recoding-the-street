# Обратотка данных
Меняем лишние символы и распологаю символы по длинее
это нужно для ситуации распространенных ситуаций ""мск г"" >>> ""к г"". Если отсортировать ""г мск"" >>> ""г м""
В базе я брал значеняи shortname именно в порядке ""г мск""

```python
stop_symbs = ['"', '.', ',', '(', ')', '№', '?']
def drop_num_empty_tab(text):
#     проверяем, есть ли что-кроме пробельных символов
    text = str(text)
    defective  = re.search("\S", text)
#     проверяем, есть ли что-то кроме цифр. (\D - любой символ кроме цифры)
    defective_1  = re.search("\D", text)
    if not defective or not defective_1 :
        return np.nan
    text = text.lower().replace('ё', 'е')
    text = text.replace('переулок', 'пер').replace('улица', 'ул')\
                .replace('гор', 'г').replace('пос', 'п').replace('район', 'р-он').replace('год', 'г')
    for symb in stop_symbs:
        text = text.replace(symb, ' ')
    list_text = text.split()
    list_text_sort = []
    for line in sorted(list_text, key=len):   
        list_text_sort.append(line)
    text = ' '.join(list_text_sort)
    return text
```

```python
# Преобразование
df['id_fz213'] = df.ID
df['ULIC_cler'] = df.ULIC.apply(drop_num_empty_tab)
df['NASP_cler'] = df.NASP.apply(drop_num_empty_tab)
df['join_s6_id'] = df.L_JION_ID_s6
```
